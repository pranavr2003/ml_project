```python
names_file = open('online_news_pop/OnlineNewsPopularity.names', 'r')
names = names_file.read()
# names
```

```python
# From names, read only the attribute names and store them in a list

names_list = names.split('\n')
names_list = names_list[41:111]
names_list = [name.split(':')[0] for name in names_list]

# names_list[i][8:] is the name of the (i+1)th attribute, starting 
# with the serial number and omitting the space before that
```

```python
# Removing the unrequired spaces and serial numbers from the attribute names

names_list2 = [name for name in names_list if name.startswith('                                       ') == False]

len(names_list2)

for name, i in zip(names_list2, range(len(names_list2))):
    names_list2[i] = name[8:]

names_list2
```
